```python
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
colors = ('r', 'g', 'b')
plt.figure(figsize=(8,5))
for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])

plt.title("Color Histogram (Feature Representation)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

filename = "E:\Sem 7\LP1\ISR\ASS 5 feature\download.png"

# Step 2: Read the image using OpenCV
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for proper display

# Step 3: Display the original image
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Step 4: Extract color features using histogram (for R, G, B channels)
colors = ('r', 'g', 'b')
plt.figure(figsize=(8,5))
for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])

plt.title("Color Histogram (Feature Representation)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

# Step 5: Print mean color values (simple numeric features)
mean_colors = cv2.mean(img)[:3]
print(f"Mean R: {mean_colors[0]:.2f}, G: {mean_colors[1]:.2f}, B: {mean_colors[2]:.2f}")
```
